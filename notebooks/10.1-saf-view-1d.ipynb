```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


csv_path = 'reports/figures/icd_on_mu.csv'

df = pd.read_csv(csv_path)
seeds = df['seed'].unique()
arr = np.array([df[df['seed'] == seed]['icd0'] for seed in seeds])
icd_means = arr.mean(axis=0)
icd_vars = arr.var(axis=0)
mus = df[df['seed']==seeds[0]]['mu']

plt.plot(mus, icd_means, 'b')
plt.fill_between(mus, icd_means-icd_vars**0.5, icd_means+icd_vars**0.5, color='lightsteelblue', alpha=0.6)    
```
